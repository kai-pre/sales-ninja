```python
DimCurrency= pd.read_csv("raw_data/DimCurrency.csv", usecols=['CurrencyKey','CurrencyName'])
DimCurrency.head()
```

```python
DimDate = pd.read_csv("raw_data/DimDate.csv", usecols=['DateKey', 'CalendarYear', 'CalendarMonthLabel', 'MonthNumber', 'CalendarQuarterLabel', 'CalendarDayOfWeekNumber', 'CalendarDayOfWeekLabel' ])
DimDate.head()
```

```python
DimGeography = pd.read_csv("raw_data/DimGeography.csv", usecols=['GeographyKey', 'ContinentName'])
DimGeography.head()
```

```python
DimProduct = pd.read_csv("raw_data/DimProduct.csv", usecols=['ProductKey', 'ProductName'])
DimProduct.head()
```

```python
DimProductCategory = pd.read_csv("raw_data/DimProductCategory.csv", usecols=['ProductCategoryKey','ProductCategoryName'])
DimProductCategory.head()
```

```python
DimProductSubcategory = pd.read_csv("raw_data/DimProductSubcategory.csv", usecols=['ProductSubcategoryKey', 'ProductSubcategoryName', 'ProductCategoryKey'])
DimProductSubcategory.head()
```

```python
DimScenario = pd.read_csv("raw_data/DimScenario.csv", usecols=['ScenarioKey', 'ScenarioName' ])
DimScenario.head()
```

```python
DimStore = pd.read_csv("raw_data/DimStore.csv", usecols=['StoreKey', 'GeographyKey', 'StoreType', 'StoreName'])
DimStore.head()
```

```python
FactSales = FactSales.merge(DimChannel, left_on='channelKey', right_on='ChannelKey', how='left')
FactSales.head(2)
```

```python
FactSales = FactSales.merge(DimCurrency, on='CurrencyKey', how='left')
FactSales.head(2)
```

```python
FactSales = FactSales.merge(DimDate, on='DateKey', how='left')
FactSales.head(2)
```

```python
DimStore = DimStore.merge(DimGeography, on='GeographyKey', how='left')
FactSales = FactSales.merge(DimStore, on='StoreKey', how='left')
FactSales.head(2)
```
